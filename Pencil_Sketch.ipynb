# Image to pencil sketch

```python
import cv2
```

Import library - OpenCV and cv2_imshow to display image in Google colab

```python
from google.colab.patches import cv2_imshow
```

Step 1 : Read the image

```python
image = cv2.imread("peacock.jpg")
cv2_imshow( image)
cv2.waitKey(0)
```

Step 2 : Create a new image by converting the original image to greyscale

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
cv2.waitKey(0)
```

Step 3 : Invert the new grayscale image

```python
inverted_image = 255 - gray_image
cv2_imshow(inverted_image)
cv2.waitKey()
```

Step 4 : Blur the image by using the Gaussian Function in OpenCV:



```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
```

Step 5 :  Invert the blurred image

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
cv2_imshow( pencil_sketch)
cv2.waitKey(0)
```

 **Original Image Vs Pencil Sketch**

```python
cv2_imshow(image)
cv2_imshow(pencil_sketch)
```
